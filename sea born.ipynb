<h2> SEABORN LIBRARY<h2>


<h5>What is seaborn?</h5>


1.Sea born is a python data visualization library based on matplotlib.


2.It provides a high-level interface for drawing attractive and informative statistical graphics.

<h5>Why seaborn?</h5>

It specializes in statistics visualization and is used if one has to summarize data in visualizations and also show the distribution in the data. 

<h2>Different categories of plot in Seaborn<h2>

Seaborn divides plot into the below categories – 


<h5>Relational plots</h5>: This plot is used to understand the relation between two variables.


<h5>Categorical plots</h5>: This plot deals with categorical variables and how they can be visualized.


<h5>Distribution plots</h5>: This plot is used for examining univariate and bivariate distributions


<h5>Regression plots</h5>: The regression plots in seaborn are primarily intended to add a visual guide that helps to emphasize patterns in a dataset during exploratory data analyses.


<h5>Matrix plots</h5>: A matrix plot is an array of scatterplots.


<h5>Multi-plot grids</h5>: It is an useful approach is to draw multiple instances of the same plot on different subsets of the dataset.

<h4>using seaborn with matplotlib</h4>

```python
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("C:\\Users\\sweth\\Desktop\\dhv\\Iris.csv")
```

```python
sns.lineplot(x="SepalWidthCm",y="SepalLengthCm",data=df)
plt.title("plotting using matplotlib function")
plt.show()
```

<h3>Changing figure Aesthetic</h3>

we use set_style() to set the aesthetic of the plot.There are five themes available in Seaborn

1.darkgrid
2.whitegrid
3.dark
4.white
5.ticks

syntax:
set_style(style=None,rc=None)

```python
sns.lineplot(x="SepalLengthCm",y="SepalWidthCm",data=df)
sns.set_style("white")
plt.show()
```

<h3>Changing the figure size</h3>

The figure size can be changed using the figure() method of matplotlib.

```python
#changing the figure size
plt.figure(figsize=(2,4))
sns.lineplot(x="SepalLengthCm",y="SepalWidthCm",data=df)
```

<h3> Scaling the plots</h3>

it can be done using the set_context() method.It allows us to override default parameters.This affects things like the the size of the labels,lines and other elements of the plot.The base context is"notebook"
and the other contexts are "paper","talk","poster"

```python
sns.lineplot(x="SepalLengthCm",y="SepalWidthCm",data=df)
sns.set_context("talk")
```

Seaborn also provide some functions for plotting multiple plots.

The functions are:

1.FacetGrid()

2.PairGrid()

FacetGrid class helps in  visualizing distribution of one variable as well as the relationship between multiple variables seperately within subsets of your dataset using multiple panels

```python
data=sns.FacetGrid(df,col="Species")
data.map(plt.plot,"SepalWidthCm")
```

Subplot grid for plotting pairwise relationships in a dataset



```python
plot=sns.PairGrid(df)
plot.map(plt.plot)
```

<h3>CountPlot</h3>


Count plot
A count plot bascially counts the categories and returns a count of their occurences.

syntax:countplot([x,y,hue,data,order,...])

```python
sns.countplot(x='Species',data=df)
```

```python
sns.set_style("dark")
sns.countplot(x='Species',palette='plasma',data=df)
```

Boxplot 

A boxplot is sometimes known as the box and whisker plt.It shows the distribution of the quantitative data that represents the comparisons between variables.

syntax:boxplot([x,y,hue,data,order,hue_order,..])

```python
sns.boxplot(x='Species',y="SepalWidthCm",data=df)
```

```python
sns.boxplot(x='Species',y="SepalWidthCm",data=df,hue="Species",palette='spring')
```

Violin Plot

It is similar to boxplot except that it provides a higher.more advanced visualization and uses the kernel density

syntax:violinplot([x,y,hue,data,order,..])

```python
sns.violinplot(x="Species",y='SepalLengthCm',data=df)
```

```python
sns.violinplot(x="Species",y='SepalLengthCm',data=df,hue='Species')
```

Stripplot

It basically creates a scatter plot based on the category.

syntax:([x,y,hue,data,order,..)]

```python
sns.stripplot(x="Species",y="SepalLengthCm",data=df)
```

```python
sns.stripplot(x="Species",y="SepalLengthCm",data=df,jitter=True,hue="Species")
```

Swarm plot

Swarm plot is very similar to the stripplot except the fact that the points are adjusted so that they do not overlap.

swarmplot([x,y,hue,data,order,..])

```python
sns.swarmplot(x='Species',y="SepalLengthCm",data=df)
```

```python
sns.swarmplot(x='Species',y="SepalLengthCm",data=df,hue='Species',palette='Oranges')
```

```python
sns.violinplot(x="Species",y='SepalLengthCm',data=df)
sns.swarmplot(x='Species',y="SepalLengthCm",data=df,color='black')
```

Factor Plot

Factor plot is the most general of all the plots and provides a parameter call kind to choose the kind of plot we want

syntax:scatterplot.factorplot([x,y,hue,data,row,col,..])

```python
sns.factorplot(x='Species',y='SepalWidthCm',data=df)
plt.show()
```

```python
sns.factorplot(x='Species',y='SepalWidthCm',data=df,kind="bar",palette='BuGn_r')
```

```python
sns.set_style('dark')
sns.factorplot(x='Species',y='SepalWidthCm',data=df,kind="violin",palette='magma')
```

```python
sns.factorplot(x='Species',y='SepalWidthCm',data=df,kind="box",palette='PuBu')
```

```python
sns.factorplot(x='Species',y='SepalWidthCm',data=df,kind="strip",palette='copper')
```

```python
sns.factorplot(x='Species',data=df,kind="count",palette='Pastel1')
```
